# Mallows 4 last 

```python
from math import prod
import sympy
import numpy as np
from actinvoting import *
```

```python
m = 4
phi = sympy.Rational(50, 100)
rho = - np.log(float(phi))
rho
```

```python
tau = [-2*rho, -rho, 0, 0]
```

```python
culture = CultureMallows(m=m, phi=phi)
session = WorkSession(culture=culture, c=m-1, tau=tau)
rate = session.p_of_zeta / prod([session.zeta[j]**(1/2) for j in session.adversaries])
float(rate)
```

## Runtimes

```python
%%timeit
equivalent_probas = equivalent_batch(session=session, ns=[25], n_jobs=1, force_recompute=True)
```

```python
exact_probas = exact_batch(session=session, ns=[25], n_jobs=1, force_recompute=True)
```

```python
monte_carlo_probas = monte_carlo_batch(session=session, ns=[25], n_samples=10000, n_jobs=1, force_recompute=True)
```
